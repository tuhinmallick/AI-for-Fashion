This notebook shows how to use [GuideLLM](https://github.com/neuralmagic/guidellm) to simulate inference workloads. The examples use a vLLM server for Qwen2-1.5B.

Note: I don't recommend running the following commands in a notebook. Instead, run them in a terminal. This way, you can monitor the server logs while GuideLLM is running and spot potential errors.

#Setup vLLM and GuideLLM

```python
!pip install vllm guidellm
```

#Start the vLLM server



```python
!nohup vllm serve "Qwen/Qwen2-1.5B" &
```

# Run GuideLLM

Before running this cell, I recommend to wait several minutes to make sure the server is properly started before running GuideLLM.


```python
!guidellm \
  --target "http://localhost:8000/v1" \
  --model "Qwen/Qwen2-1.5B" \
  --data-type emulated \
  --data "prompt_tokens=512,generated_tokens=128"
```
