This notebook shows how to quantize and run Llama 3 70B with mixed precision using ExLlamaV2.

It requires at least 16 GB of GPU RAM for the measurement step.

```python
!pip install --upgrade transformers exllamav2 datasets
!git clone https://github.com/turboderp/exllamav2.git
```

ExLlamaV2 doesn't communicate well with Hugging Face transformers. You first need to download the models locally. We only need the safetensors version. ExLlamaV2 uses safetensors so we don't need to download the ".bin" files.

```python
#The directory where we want to store the model must exist.
!mkdir ./Llama-3-70B/
```

```python
from huggingface_hub import snapshot_download
snapshot_download(repo_id="meta-llama/Meta-Llama-3-70B", ignore_patterns=["*.bin", "*.pth"], local_dir="./Llama-3-70B/", local_dir_use_symlinks=False)
```

The first step measures the quantization errors for different quantization configurations. It takes several hours.

```python
!mkdir ./Llama-3-70B/temp/
!python exllamav2/convert.py \
    -i ./Llama-3-70B/ \
    -o ./Llama-3-70B/temp/ \
    -nr \
    -om ./Llama-3-70B/measurement.json
```

Then, we can quantize the model as follows (example for a 4-bit quantization):

```python
!python exllamav2/convert.py \
    -i ./Llama-3-70B/ \
    -o ./Llama-3-70B/temp/ \
    -nr
    -m  ./Llama-3-70B/measurement.json\
    -cf ./Llama-3-70B/4.0bpw/ \
    -b 4.0
```

You can use test_inference.py to try the model. For the demonstration, I use Llama 3 70B with a 2.5 bpw.

To evaluate the perplexity with the script, I use the test split of the openassistant-guanaco. It needs to be in the parquet format.

```python
!wget -O oasstguanaco-test.parquet https://huggingface.co/datasets/timdettmers/openassistant-guanaco/resolve/refs%2Fconvert%2Fparquet/default/test/0000.parquet?download=true
```

```python
!python exllamav2/test_inference.py -m ./Llama-3-70B/2.5bpw/ --eval_dataset oasstguanaco-test.parquet --eval_rows 518 --eval_length 256 -p "Once upon a time," --tokens 512
```
